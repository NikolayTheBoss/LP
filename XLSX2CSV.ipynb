```python
import pandas as pd
import re

company_regex = re.compile('ООО.*[\<"].*["\>]|ИП\s[а-яА-Я]*[а-яА-Я]*\s*[а-яА-Я]*\s*[а-яА-Я]*')
phone_regex = re.compile('8\s\(\d{3,4}\)\s[\d-]{8,9}|\+7\s\d{3}\s[\d-]{9}')
inn_regex = re.compile('\d{10}|\d{12}')
email_regex = re.compile('[\.\-0-9a-zA-Z]*@[\.\-0-9a-zA-Z]*\.[\.\-0-9a-zA-Z]{1,5}')
webstie_regex = re.compile('https{0,1}\:\/\/[w]{0,3}\.{0,1}.*\.[a-zA-Zа-яА-Я]{1,5}')
export_list = []
df = pd.read_excel('3000.xlsx', names=['', '', 'main_data', 'inn', 'kod'])
for i in df.main_data:
    if str(i) != 'nan':
        
        company_name = ''.join(re.findall(company_regex, i)).replace('<','"').replace('>','"')
        phone_number = ','.join(re.findall(phone_regex, i))
        inn_number = ''.join(re.findall(inn_regex, i))
        email = ''.join(re.findall(email_regex, i))
        website = ''.join(re.findall(webstie_regex, i))
        
        export_list.append({'Название':company_name,'ИНН':inn_number,'Телефон':phone_number,'Почта':email,'Сайт':website})
```

```python
import csv

with open('output.csv', 'w') as file:
    writer = csv.DictWriter(file, delimiter='|', fieldnames=['Название', 'ИНН', 'Телефон', 'Почта', 'Сайт'])
    writer.writeheader()
    for company in export_list:
        writer.writerow(company)
print('Done!')
```
